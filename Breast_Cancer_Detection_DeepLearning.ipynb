# **BREAST CANCER CLASSIFICATION USING NEURAL NETWORK**

**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, confusion_matrix)
```

**Data Collection & Processing**

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
print("Dataset shape:",data_frame.shape)
```

```python
# getting some information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Target Varibale
print("Target distribution:\n", data_frame['label'].value_counts())
plt.figure(figsize=(6,4))
sns.countplot(x=data_frame['label'])
plt.title('Class Distribution')
plt.show()
```

**1** **-->** **Benign**

**0** **-->** **Malignant**

```python
data_frame.groupby('label').mean()
```

**Separating the features and target**

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting the data into training data & Testing data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

**Building the Neural Network**

download.png

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

**Visualizing accuracy and loss of train data**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

**Accuracy of the model on test data**

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
print(X_test_std.shape)
print(X_test_std[0])
```

```python
Y_pred_prob = model.predict(X_test_std)
```

```python
print(Y_pred_prob.shape)
print(Y_pred_prob[0])
```

```python
print(X_test_std)
```

```python
print(Y_pred_prob)
```

**model.predict() gives the prediction probability of each class for that data point**

```python
# converting the prediction probability to class labels

Y_pred_labels = [np.argmax(i) for i in Y_pred_prob]
print(Y_pred_labels)
```

```python
#Performance of the model
dl_metrics = {
    'Accuracy': accuracy_score(Y_test, Y_pred_labels),
    'Precision': precision_score(Y_test, Y_pred_labels),
    'Recall': recall_score(Y_test, Y_pred_labels),
    'F1-score': f1_score(Y_test, Y_pred_labels),
    'ROC AUC': roc_auc_score(Y_test, Y_pred_prob[:, 1])
}
print("Deep Learning model performance:\n", dl_metrics)

# Confusion matrix
cm = confusion_matrix(Y_test, Y_pred_labels)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

**Building the predictive system**

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

input_df = pd.DataFrame(input_data_reshaped, columns=scaler.feature_names_in_)
# standardizing the input data
input_data_std = scaler.transform(input_df)

prediction = model.predict(input_data_std)
print("Probability of the tumour being Malignant or Benign is:", prediction)

prediction_label = [np.argmax(prediction)]

classes = ['Malignant', 'Benign']
predicted = classes[prediction_label[0]]


print("\n" + "="*40)
print("           PREDICTION SUMMARY           ")
print("="*40)
print(f" Tumor Type : {predicted}")
```
